```python
import xarray as xr
from glob import glob
from virtualizarr import open_virtual_dataset
```

```python
#file all *.nc files in the data folder recursively

files = glob('data/*.nc')
files = sorted(files)
```

```python
list_of_virtual_datasets = [open_virtual_dataset(f) for f in files]
list_of_virtual_datasets
```

```python
temp_ds = list_of_virtual_datasets[0]

for ds in list_of_virtual_datasets[1:]:
    temp_ds = xr.concat([temp_ds, ds], dim='time')

virtual_ds = temp_ds
```

```python
virtual_ds.virtualize.to_kerchunk('x.json', format='json')
```

```python
combined_ds = xr.open_dataset('x.json', engine="kerchunk")
```
